````python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def delta_batch(W, X, D):
    alpha = 0.9
    
    dWsum = np.zeros((3, 1))
    
    N = 4
    
    for k in range(N):
        x = X[k, :].reshape(-1, 1)
        d = D[k]
        
        v = np.dot(W, x)
        y = sigmoid(v)
        
        e = d - y
        delta = y * (1 - y) * e
        
        dW = alpha * delta * x
        
        dWsum += dW
    
    dWavg = dWsum / N
    
    W[0] += dWavg[0, 0]
    W[1] += dWavg[1, 0]
    W[2] += dWavg[2, 0]
    
    return W
````

# 과적합 overfitting 문제
- 정직화 regularization : 모델의 형태를 최대한 간단하게 만들려는 수치해석적인 기법
- 검증 validation : 학습 데이터의 일부를 떼어내서 학습에 사용하지 않고 모델 성능 검증용으로 사용하는 기법 곧 학습된 모델에 검증 데이터를 입력해 성능 저하가 심하면 과적합으로 보고 모델을 수정하는 등의 조치를 취해 과적합을 방지
- 학습데이터를 학습용과 검증용으로 분류 8:2 로 나누어 학습용 데이터로 모델을 학습 후 검증용 데이터로 모델 성능 평가, 성능 만족하면 종료, 그렇지 않으면 모델 구조 등을 수정해 2단계 수행
- 교차 검증 cross validation : 학습용과 검증용으로 나누는 것은 같은데, 고정하지 않고 계속 바꿔준다는 것이 다름. 예로, 중간중간 학습용, 검증용 데이터를 랜덤하게 나누어 학습 진행
